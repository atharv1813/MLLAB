```python
import pandas as pd
import numpy as np

data = pd.read_csv('car_evaluation.csv')
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
data.rename(columns={'vhigh': 'buyprice', 'vhigh.1': 'maintainprice', '2': 'doors', '2.1': 'persons', 'small': 'luggagesize', 'low': 'safety', 'unacc':'rating'}, inplace=True)
data
```

```python
from scipy.stats import chi2_contingency

features = data.columns.tolist()

for i in range(len(features)):
    for j in range(i + 1, len(features)):
        contingency_table = pd.crosstab(data[features[i]], data[features[j]])
        chi2, p, _, _ = chi2_contingency(contingency_table)
        print(f"Chi-Square Test of Independence between {features[i]} and {features[j]}:")
        print(f"Chi-Square Value: {chi2}")
        print(f"P-value: {p}")
        if p < 0.05:  
            print("There is a significant relationship between the two variables.")
        else:
            print("The two variables are independent.")
        print("\n")
```

```python
def train_test_split(data, test_size=0.2):
    np.random.seed(99)  # 99 la jast yetay 89 chya aaspaas
    shuffled_indices = np.random.permutation(len(data))
    test_set_size = int(len(data) * test_size)
    test_indices = shuffled_indices[:test_set_size]
    train_indices = shuffled_indices[test_set_size:]
    return data.iloc[train_indices], data.iloc[test_indices]
```

```python
train_data, test_data = train_test_split(data, test_size=0.2)

X_train = train_data.drop(columns=['rating'])
Y_train = train_data['rating']
X_test = test_data.drop(columns=['rating'])
Y_test = test_data['rating']
```

```python
unique_ratings = data['rating'].unique()
class_priors = {}
for rating in unique_ratings:
    count = (data['rating'] == rating).sum()  
    prob = count / data.shape[0]
    class_priors[rating] =  prob
```

```python
conditional_probs = {}
for feature in X_train.columns:
    feature_probs = {}
    for category in X_train[feature].unique():
        for label in Y_train.unique():
            count_feature_and_label = len(train_data[(X_train[feature] == category) & (Y_train == label)])
            count_label = len(Y_train[Y_train == label])
            prob = (count_feature_and_label + 1) / (count_label + len(X_train[feature].unique()))
            feature_probs[(category, label)] = prob
    conditional_probs[feature] = feature_probs
```

```python
correct_predictions = 0
posterior_probs = {label: np.log(class_priors[label]) for label in Y_train.unique()}
predictions = []

for i in range(len(test_data)):
    row = test_data.iloc[i]
    
    for feature in X_train.columns:
        for label in Y_train.unique():
            conditional_prob = conditional_probs[feature].get((row[feature], label))
            if conditional_prob is None:
                conditional_prob = 1e-10

            likelihood = np.log(conditional_prob)
            posterior_probs[label] += likelihood

    predicted_label = None
    max_prob = float('-inf')

    for label, prob in posterior_probs.items():
        if prob > max_prob:
            max_prob = prob
            predicted_label = label

    if predicted_label == row['rating']:
        correct_predictions += 1
    
    predictions.append(predicted_label)
    posterior_probs = {label: np.log(class_priors[label]) for label in Y_train.unique()}


print(posterior_probs)
```

```python
accuracy = correct_predictions / len(test_data) * 100
print(f"Accuracy of my model on the test set is: {accuracy:.2f}%")
```

```python
from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns
conf_matrix = confusion_matrix(Y_test, predictions)
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```
