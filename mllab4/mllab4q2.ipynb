```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler

data = pd.read_csv('iris.csv')
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
data
 
```

```python
data.isnull().sum()
```

```python
for column in data.columns:
    if data[column].dtype == 'object':
        unique_values = data[column].unique()
        print(f"Column '{column}' has {len(unique_values)} unique values:")
        print(unique_values)
        print("\n")
    else:
        min_value = data[column].min()
        max_value = data[column].max()
        print(f"Column '{column}' has a range from {min_value} to {max_value}")
        print("\n")
```

```python
X = data[['sepal_length', 'sepal_width', 'petal_length', 'petal_width']]
Y = data['species']
```

```python
Y_m1 =  data['species'].apply(lambda x: 1 if x == 'setosa' else 0)
Y_m2=  data['species'].apply(lambda x: 1 if x == 'versicolor' else 0)
Y_m3 =  data['species'].apply(lambda x: 1 if x == 'virginica' else 0)
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_scaled
```

```python
import math

def sigmoid(z):
    return 1 / (1 + np.exp(-z))

class LogisticRegression:

    def __init__(self, learning_rate=0.01, num_iterations=1000):
        self.learning_rate = learning_rate
        self.num_iterations = num_iterations
        self.w = None
        self.b = None

    def fit(self, X, Y):
        num_samples, num_features = X.shape
        self.w = np.zeros(num_features)
        self.b = 0

        for _ in range(self.num_iterations):
            linear_pred = np.dot(X, self.w) + self.b
            predictions = sigmoid(linear_pred)

            dw = (1/num_samples) * np.dot(X.T, (predictions - Y))
            db = (1/num_samples) * np.sum(predictions - Y)

            self.w -= self.learning_rate * dw
            self.b -= self.learning_rate * db

    def predict(self, X):
        linear_pred = np.dot(X, self.w) + self.b
        y_pred = sigmoid(linear_pred)
        
        return y_pred   # bas ye change hus pichle wale se kyu dekhlo niche jake
```

```python
split_ratio = 0.8
split_idx = int(len(X_scaled) * split_ratio)
np.random.seed(20)
indices = np.arange(len(X_scaled))
np.random.shuffle(indices)
#print(indices)
X_shuffled = X_scaled[indices]
Y_shuffled = Y[indices]
Y_m1_shuffled, Y_m2_shuffled, Y_m3_shuffled = Y_m1[indices], Y_m2[indices], Y_m3[indices]

X_train, X_test = X_shuffled[:split_idx], X_shuffled[split_idx:]
Y_train_m1, Y_train_m2, Y_train_m3 = Y_m1_shuffled[:split_idx], Y_m2_shuffled[:split_idx], Y_m3_shuffled[:split_idx]
Y_test_m1, Y_test_m2, Y_test_m3 = Y_m1_shuffled[split_idx:], Y_m2_shuffled[split_idx:], Y_m3_shuffled[split_idx:]
Y_test_final = Y_shuffled[split_idx:]

lr_m1 = LogisticRegression(learning_rate=0.01, num_iterations=1000)
lr_m2 = LogisticRegression(learning_rate=0.01, num_iterations=1000)
lr_m3 = LogisticRegression(learning_rate=0.01, num_iterations=1000)

lr_m1.fit(X_train, Y_train_m1)
lr_m2.fit(X_train, Y_train_m2)
lr_m3.fit(X_train, Y_train_m3)

test_predictions_m1 = lr_m1.predict(X_test)
test_predictions_m2 = lr_m2.predict(X_test)
test_predictions_m3 = lr_m3.predict(X_test)
```

```python
ans = []

for i, j, k in zip(test_predictions_m1, test_predictions_m2, test_predictions_m3):
    if i > j and i > k:
        ans.append('setosa')
    elif j > i and j > k:
        ans.append('versicolor')
    elif k > i and k > j:
        ans.append('virginica')
```

```python
num_correct = sum(1 for true_label, predicted_label in zip(Y_test_final, ans) if true_label == predicted_label)

total_samples = len(Y_test_final)
accuracy = num_correct / total_samples

accuracy_percentage = accuracy * 100

print("Accuracy on Test Data: {:.2f} %".format(accuracy_percentage))
```

```python
from sklearn.metrics import precision_score, recall_score, f1_score

precision = precision_score(Y_test_final, ans, average='weighted')
recall = recall_score(Y_test_final, ans, average='weighted')
f1 = f1_score(Y_test_final, ans, average='weighted')

print("Precision: {:.2f}".format(precision))
#Precision = TP / (TP + FP)
print("Recall: {:.2f}".format(recall))
#Recall = TP / (TP + FN)
print("F1-Score: {:.2f}".format(f1))
#F1 Score = 2 * (Precision * Recall) / (Precision + Recall)
```

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

cm = confusion_matrix(Y_test_final, ans)

plt.figure(figsize=(8, 6))
sns.set(font_scale=2)
sns.heatmap(cm, annot=True, fmt='g', cmap="Blues", cbar=True,
            annot_kws={"size": 16}, linewidths=0.5)
plt.xlabel('Predicted labels')
plt.ylabel('True labels')
plt.title('Confusion Matrix')
plt.show()
```

```python
import numpy as np

num_folds = 5
best_fold = None
best_accuracy = 0.0
accuracies = []


num_samples = len(X_scaled)
fold_size = num_samples // num_folds

indices = np.arange(num_samples)
np.random.seed(20)
np.random.shuffle(indices)

for fold in range(num_folds):
    start_idx = fold * fold_size
    end_idx = (fold + 1) * fold_size

    val_indices = indices[start_idx:end_idx]
    train_indices = np.concatenate([indices[:start_idx], indices[end_idx:]])

    X_train, X_val = X_scaled[train_indices], X_scaled[val_indices]
    Y_train_m1, Y_train_m2, Y_train_m3 = Y_m1.iloc[train_indices], Y_m2.iloc[train_indices], Y_m3.iloc[train_indices]
    Y_val_final = Y.iloc[val_indices]

    lr_m1 = LogisticRegression(learning_rate=0.01, num_iterations=1000)
    lr_m2 = LogisticRegression(learning_rate=0.01, num_iterations=1000)
    lr_m3 = LogisticRegression(learning_rate=0.01, num_iterations=1000)

    lr_m1.fit(X_train, Y_train_m1)
    lr_m2.fit(X_train, Y_train_m2)
    lr_m3.fit(X_train, Y_train_m3)
    
    test_predictions_m1 = lr_m1.predict(X_val)
    test_predictions_m2 = lr_m2.predict(X_val)
    test_predictions_m3 = lr_m3.predict(X_val)

    ans = []

    for i, j, k in zip(test_predictions_m1, test_predictions_m2, test_predictions_m3):
        if i > j and i > k:
            ans.append('setosa')
        elif j > i and j > k:
            ans.append('versicolor')
        elif k > i and k > j:
            ans.append('virginica')

    num_correct = sum(1 for true_label, predicted_label in zip(Y_val_final, ans) if true_label == predicted_label)
    total_samples = len(Y_val_final)
    accuracy = num_correct / total_samples
    accuracies.append(accuracy)

    if accuracy > best_accuracy:
        best_accuracy = accuracy
        best_fold = fold

average_accuracy = np.mean(accuracies)
print("Average Accuracy on Validation Data ({}-fold cross-validation): {:.2f} %".format(num_folds, average_accuracy * 100))
print("Best Fold:", best_fold)
print("Best Accuracy:", best_accuracy * 100)
```
