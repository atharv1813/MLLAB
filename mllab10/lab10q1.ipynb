```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def threshold_activation(x):
    return 1 if x >= 0 else 0
```

```python
def train_perceptron(inputs, targets, learning_rate, epochs, weight_min=-0.3, weight_max=0.3):
    num_inputs = len(inputs[0])
    num_samples = len(inputs)

    weights = np.random.uniform(weight_min, weight_max, num_inputs)

    training_accuracy = []

    for epoch in range(epochs):
        correct_predictions = 0
        for i in range(num_samples):
            input_data = inputs[i]
            target = targets[i]

            weighted_sum = np.dot(weights, input_data)

            prediction = threshold_activation(weighted_sum)

            error = target - prediction
            weights += learning_rate * error * input_data

            if error == 0:
                correct_predictions += 1

        accuracy = correct_predictions / num_samples
        training_accuracy.append(accuracy)

    return weights, training_accuracy
```

```python
or_inputs = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
or_targets = np.array([0, 1, 1, 1])

and_inputs = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
and_targets = np.array([0, 0, 0, 1])

learning_rates = [0.1, 0.3, 0.5, 0.7, 0.9]
epochs = 100
```

```python
lr = learning_rates[0]
print(f"Training for learning rate {lr}...")
_, training_accuracy_or = train_perceptron(or_inputs, or_targets, lr, epochs)
_, training_accuracy_and = train_perceptron(and_inputs, and_targets, lr, epochs)

plt.figure()
plt.plot(range(epochs), training_accuracy_or, label="OR Problem")
plt.plot(range(epochs), training_accuracy_and, label="AND Problem")
plt.title(f"Training Curve for Learning Rate = {lr}")
plt.xlabel("Epochs")
plt.ylabel("Training Accuracy")
plt.legend()
plt.show()
```

```python
Why is zero initialization of weight, not a good initialization technique?
If we initialize the set of weights in the neural network as zero, then all the neurons at each layer will
start producing the same output and the same gradients during backpropagation.
As a result, the neural network cannot learn anything at all because there is no source of asymmetry
between different neurons. Therefore, we add randomness while initializing the weight in neural networks
```

```python
Here are the common activation functions that can cause the vanishing gradient problem in deep neural networks:

- Sigmoid - The sigmoid squashes real-valued inputs to between 0 and 1.Its gradient approaches 0 at extremes leading to vanishing gradients.

- Tanh - Like sigmoid, the tanh function also saturates at -1 and 1, with gradients diminishing to 0 at extremes. 

- Softplus - This is a smooth approximation to ReLU. But it saturates at high values, reducing slopes.

- ELU - The Exponential Linear Unit improves on ReLUs but still has vanishing gradients on the negative side.

- Softmax - While useful for probabilities, softmax flattens for large inputs causing small gradients.

- Gaussian Error Linear Units (GELU) - Approximates Gaussian distribution but gradients vanish at extremes.

- Bent Identity - Squashes inputs between -1 and 1 similar to tanh, leading to saturation.

- Arctan - Has vanishing gradients when inputs approach +/- infinity due to horizontal asymptotes. 

- Hard sigmoid - Piecewise linear approximation to sigmoid but still saturates at 0 and 1.

- Hard tanh - Bounded between -1 and 1 with flat gradients at extremes.

So in essence, any activation function that compresses real-valued inputs into a finite range can potentially 
exhibit vanishing gradients at the extremes. The key exceptions are ReLU and its variants like Leaky ReLU and PReLU,
which avoid this by not saturating.
```

```python
Here is an explanation of some common activation functions used in artificial neural networks:

Sigmoid - It takes a real-valued input and squashes it to range between 0 and 1. Mathematically, it is represented as
σ(x) = 1 / (1 + e-x). The curve is S-shaped with smooth gradient. However, it suffers from vanishing gradient problem for 
very high or low values.

Tanh - Tanh or hyperbolic tangent is similar to sigmoid but outputs values between -1 and 1. Mathematically, it is given as 
tanh(x) = (e2x - 1) / (e2x + 1). Like sigmoid, it also has a smooth gradient but is zero-centered. The slope is steeper 
compared to sigmoid.

ReLU - Rectified Linear Unit applies a threshold at zero and outputs the input if positive, else outputs zero. It is 
represented as f(x) = max(0, x). Computationally efficient and avoids vanishing gradient problem. But dead neurons can 
occur when gradients reach zero.

Leaky ReLU - This addresses the dead neuron problem in ReLU by allowing a small negative slope (alpha). Represented as
f(x) = max(alpha*x, x). Alpha is usually set as 0.01 or 0.1.

Softmax - Converts output layer values into probability distributions summing to 1. Widely used in multiclass classification. 
Denoted as σ(zi) = ezi / Σj ezj where z is the vector of outputs from the final layer.

In summary, sigmoid and tanh squash real-valued inputs between ranges but have vanishing gradients, ReLUs introduce 
non-linearity while avoiding vanishing gradient but have dead neuron issues, and softmax provides probability distribution 
over output classes.
```

```python
18. Why is ReLU the most commonly used Activation Function?
ReLU (Rectified Linear Unit) is the most commonly used activation function in neural networks due to the following reasons:

1. No vanishing gradient: The derivative of the RELU activation function is either 0 or 1, so it could be not in the
    range of [0,1]. As a result, the product of several derivatives would also be either 0 or 1, because of this property,
    the vanishing gradient problem doesn’t occur during backpropagation.

2. Faster training: Networks with RELU tend to show better convergence performance. Therefore, we have a much lower run time.

3. Sparsity: For all negative inputs, a RELU generates an output of 0. This means that fewer neurons of the network are
    firing. So we have sparse and efficient activations in the neural network.
```
