```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.array([0,1,2,3,1,1.5])
Y = np.array([1,2,1.5,2,1,2])

print("X:", "  ".join(f"{x_val:2f}" for x_val in X))
print("Y:", "  ".join(f"{y_val:2f}" for y_val in Y))
```

```python
A = np.zeros((3, 3))
B = np.zeros(3)

for xi, yi in zip(X, Y):
    A[0, 0] += 1
    A[0, 1] += xi
    A[0, 2] += xi ** 2
    A[1, 0] += xi
    A[1, 1] += xi ** 2
    A[1, 2] += xi ** 3
    A[2, 0] += xi ** 2
    A[2, 1] += xi ** 3
    A[2, 2] += xi ** 4

    B[0] += yi
    B[1] += xi * yi
    B[2] += xi ** 2 * yi
```

```python
x = np.linalg.solve(A, B)
```

```python
print("Coefficients are:")
print("b0 =", x[0])
print("b1 =", x[1])
print("b2 =", x[2])
```
