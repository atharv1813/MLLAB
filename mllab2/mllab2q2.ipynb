```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.random.uniform(1, 3, size=(6,))
Y = 1.7 + 0.5 * X
```

```python
print("X:", "  ".join(f"{x_val:2f}" for x_val in X))
print("Y:", "  ".join(f"{y_val:2f}" for y_val in Y))
```

```python
learning_rate = 0.1
epochs = 3
slope = 0
intercept = 0
```

```python
m_values = []
c_values = []
```

```python
for _ in range(epochs):
    predictions = slope * X + intercept
    error = predictions - Y
    slope -= learning_rate * (2 * np.dot(error, X) / len(X))
    intercept -= learning_rate * (2 * np.sum(error) / len(X))
    m_values.append(slope)
    c_values.append(intercept)
```

```python
print("\nSlope values:", m_values)
print("Intercept values:", c_values)
```

```python
Ynew = m_values[0]*X + c_values[0]
for xi, yi, Yi in zip(X,Y, Ynew):
    print(f'x: {xi}   y: {yi}   y_new: {Yi}')
```

```python
Ynew = m_values[1]*X + c_values[1]
for xi, yi, Yi in zip(X,Y, Ynew):
    print(f'x: {xi}   y: {yi}   y_new: {Yi}')
```

```python
Ynew = m_values[2]*X + c_values[2]
for xi, yi, Yi in zip(X,Y, Ynew):
    print(f'x: {xi}   y: {yi}   y_new: {Yi}')
```

```python
def calculate_cost(X, Y, m, c):
    n = len(X)
    predictions = m * X + c
    squared_errors = (predictions - Y)**2
    cost = 1 / (2 * n) * np.sum(squared_errors)
    return cost
```

```python
for i in range(epochs):
    cost = calculate_cost(X, Y, m_values[i], c_values[i])
    print(f"Iteration {i + 1} : Cost(MSE): {cost:.4f}")
```

```python
plt.scatter(X, Y, label='Data Points')
plt.plot(X, m_values[0] * X + c_values[0], color='red', label='Line 1')
plt.plot(X, m_values[1] * X + c_values[1], color='green', label='Line 2')
plt.plot(X, m_values[2] * X + c_values[2], color='yellow', label='Line 3')
plt.xlabel('X(independant feature)')
plt.ylabel('Y(dependant feature)')
plt.title('Linear Regression using Gradient Descent')
plt.legend()
plt.show()
```
