# k(num_points) is hyperparameter

```python
import numpy as np

def single_linkage(distance_matrix, cluster1, cluster2):
    return np.min(distance_matrix[np.ix_(cluster1, cluster2)])

def complete_linkage(distance_matrix, cluster1, cluster2):
    return np.max(distance_matrix[np.ix_(cluster1, cluster2)])

def average_linkage(distance_matrix, cluster1, cluster2):
    return np.mean(distance_matrix[np.ix_(cluster1, cluster2)])

def centroid_linkage(data, cluster1, cluster2):
    centroid1 = np.mean(data[cluster1], axis=0)
    centroid2 = np.mean(data[cluster2], axis=0)
    return np.linalg.norm(centroid1 - centroid2)

def create_distance_matrix(data):
    num_points = len(data)
    distance_matrix = np.zeros((num_points, num_points))
    for i in range(num_points):
        for j in range(i, num_points):
            distance = np.linalg.norm(data[i] - data[j])
            distance_matrix[i][j] = distance
            distance_matrix[j][i] = distance
    return distance_matrix

def print_clusters(clusters):
    for i, cluster in enumerate(clusters):
        num_points = len(cluster)
        print(f"Cluster {i + 1} has {num_points} points.")

def agglomerative_clustering(data, k):
    num_points = len(data)
    distance_matrix = create_distance_matrix(data)
    clusters = [list([i]) for i in range(num_points)]

    while len(clusters) > k:
        distances = [
            (i, j, centroid_linkage(distance_matrix, clusters[i], clusters[j]))
            for i in range(len(clusters))
            for j in range(i + 1, len(clusters))
        ]
        i, j, _ = min(distances, key=lambda x: x[2])

        clusters[i] = clusters[i] + clusters[j]
        del clusters[j]

    return clusters
```

```python
num_clusters = 10
clusters_list = agglomerative_clustering(datafinal2, num_clusters)

print_clusters(clusters_list)
```

```python
def plot_cluster_sizes(clusters):
    cluster_sizes = [len(cluster) for cluster in clusters]
    cluster_numbers = list(range(1, len(clusters) + 1))

    plt.bar(cluster_numbers, cluster_sizes)
    plt.xlabel('Cluster Number')
    plt.ylabel('Number of Points')
    plt.title('Cluster Sizes')
    plt.show()
```

```python
plot_cluster_sizes(clusters_list)
```

```python
cluster_sizes = [len(cluster) for cluster in clusters_list]

total_data_points = sum(cluster_sizes)

percentages = [(size / total_data_points) * 100 for size in cluster_sizes]

plt.figure(figsize=(6, 6))
plt.pie(percentages, labels=["Cluster " + str(i+1) for i in range(len(cluster_sizes))], autopct='%1.1f%%')

plt.title("Cluster Sizes")

plt.axis('equal')
plt.show()
```

```python
The provided code is an implementation of agglomerative hierarchical clustering in Python. Agglomerative hierarchical 
clustering is a bottom-up approach to clustering data, where each data point initially forms its own cluster, and clusters 
are successively merged together until a stopping criterion is met, such as a desired number of clusters (k). 

5. **Agglomerative Clustering:**
   - The `agglomerative_clustering` function performs the actual hierarchical clustering. It takes the input data and the 
      desired number of clusters (k) as parameters.
   - Initially, each data point is assigned to its own cluster. These individual data points are stored in the `clusters` list.
   - In a loop, clusters are successively merged until the number of clusters is reduced to k.
   - The code calculates the pairwise distances between all clusters using the specified linkage method and 
     stores these distances in the `distances` list.
   - It then selects the pair of clusters with the smallest distance and merges them into a single cluster.
   - The cluster that is merged is removed from the list of clusters.
   - The process continues until the desired number of clusters (k) is reached.

6. **Return Clusters:**
   - Finally, the `agglomerative_clustering` function returns the list of clusters, where each cluster is represented as
    a list of indices of data points.

In summary, this code performs agglomerative hierarchical clustering on a given dataset, starting with 
each data point in its own cluster and successively merging clusters based on the specified linkage method until
the desired number of clusters (k) is achieved. The specific linkage method used can be chosen from single, complete, 
average, or centroid linkage, depending on the desired behavior of the clustering algorithm.
```
